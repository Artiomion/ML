```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
df = pd.read_csv('housing.csv', header=None, delimiter=r"\s+", names=column_names)

print("Размер датасета:", df.shape)

'''
CRIM – уровень преступности на душу населения в городе.
ZN – доля жилых земель, отведенных под участки площадью более 25 000 кв. футов (25 000 кв. футов).
INDUS – доля некоммерческих площадей в городе.
CHAS — фиктивная переменная реки Чарльз (1, если участок граничит с рекой; 0 в противном случае)
NOX — концентрация оксидов азота (частей на 10 миллионов)
RM — среднее количество комнат в жилище
AGE — доля домов, занимаемых владельцами, построенных до 1940 года
DIS — взвешенное расстояние до пяти бостонских центров занятости
RAD — индекс доступности к радиальным магистралям
TAX — ставка налога на имущество на полную стоимость за 10 000 долларов США
PTRATIO — соотношение учащихся и учителей по городам
B — 1000(Bk — 0,63)^2, где Bk — доля чернокожего населения по городам
LSTAT — процент населения с низким социальным статусом
MEDV — медианная стоимость домов, занимаемых владельцами, в тысячах долларов США
'''
```

```python
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
column_sels = ['LSTAT', 'INDUS', 'NOX', 'PTRATIO', 'RM', 'TAX', 'DIS', 'AGE']
x = df.loc[:,column_sels]
y = df['MEDV']
x = pd.DataFrame(data=min_max_scaler.fit_transform(x), columns=column_sels)
fig, axs = plt.subplots(ncols=4, nrows=2, figsize=(20, 10))
index = 0
axs = axs.flatten()
for i, k in enumerate(column_sels):
    sns.regplot(y=y, x=x[k], ax=axs[i])
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```

```python
cols_to_clip = ['CRIM', 'TAX', 'LSTAT', 'DIS', 'B']
df_clean = df.copy()
iqr_info = {}
k = 3  # можно увеличить до 3.0 для более мягкого клиппинга

for c in cols_to_clip:
    q1, q3 = df_clean[c].quantile([0.25, 0.75])
    iqr = q3 - q1
    lo = q1 - k * iqr
    hi = q3 + k * iqr
    iqr_info[c] = (lo, hi)
    df_clean[c] = df_clean[c].clip(lower=lo, upper=hi)

print("Клиппинг выполнен для:", cols_to_clip)
print({c: (round(lo, 3), round(hi, 3)) for c, (lo, hi) in iqr_info.items()})

# Используем очищенную версию дальше по ноутбуку
df = df_clean
fig, axs = plt.subplots(ncols=7, nrows=2, figsize=(20, 10))
index = 0
axs = axs.flatten()
for k, v in df.items():
    sns.boxplot(y=k, data=df, ax=axs[index])
    index += 1
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X = df.drop(columns=["MEDV"]).to_numpy(dtype=float)
y = df["MEDV"].to_numpy(dtype=float)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

class MyLinearRegression:
    def __init__(self, lr, iterations):
        self.lr = lr
        self.iterations = iterations
        self.w = []
        self.b = 0

    def fit(self, X, y):
        m, n = len(X), len(X[0])
        self.w = [0.0] * n

        for _ in range(self.iterations):
            y_pred = [sum(self.w[j] * X[i][j] for j in range(n)) + self.b for i in range(m)]
            errors = [y[i] - y_pred[i] for i in range(m)]

            dw = [(-2/m) * sum(X[i][j] * errors[i] for i in range(m)) for j in range(n)]
            db = (-2/m) * sum(errors)

            for j in range(n):
                self.w[j] -= self.lr * dw[j]
            self.b -= self.lr * db

    def predict(self, X):
        return [sum(self.w[j] * X[i][j] for j in range(len(self.w))) + self.b for i in range(len(X))]


my_model = MyLinearRegression(lr=0.01, iterations=200000)
my_model.fit(X_train.tolist(), y_train.tolist())
pred_my = my_model.predict(X_test.tolist())

sk_model = LinearRegression()
sk_model.fit(X_train, y_train)
pred_sk = sk_model.predict(X_test)

def mse(y_true, y_pred):
    return sum((y_true[i] - y_pred[i])**2 for i in range(len(y_true))) / len(y_true)

def r2_score_percent(y_true, y_pred):
    y_mean = sum(y_true) / len(y_true)
    ss_tot = sum((y - y_mean)**2 for y in y_true)
    ss_res = sum((y_true[i] - y_pred[i])**2 for i in range(len(y_true)))
    return (1 - ss_res / ss_tot) * 100

print("MSE (моя модель, test):", mse(y_test, pred_my))
print("MSE (sklearn, test):", mse(y_test, pred_sk))

print("\nПример предсказаний (тестовые данные):")
for i in range(5):
    print(f"y={y_test[i]:.1f}, my={pred_my[i]:.1f}, sk={pred_sk[i]:.1f}")

print("\nR² (моя модель, test): {:.1f}%".format(r2_score_percent(y_test, pred_my)))
print("R² (sklearn, test): {:.1f}%".format(r2_score_percent(y_test, pred_sk)))
```
