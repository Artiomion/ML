```python
def get_title(name):
    if pd.isna(name):
        return None
    # Имя в формате "Moran, Mr. James"
    if "Mr." in name:
        return "Mr"
    elif "Mrs." in name:
        return "Mrs"
    elif "Miss." in name or "Ms." in name:
        return "Miss"
    elif "Master." in name:
        return "Master"
    else:
        return "Other"

# Создаём новый столбец с титулом
df['Title'] = df['Name'].apply(get_title)
title_age_median = df.groupby('Title')['Age'].median()

# Функция для заполнения возраста
df['Age'] = df.apply(lambda row: title_age_median[row['Title']] if pd.isna(row['Age']) else row['Age'], axis=1)
```

```python
def predict_survival(passenger: dict) -> int:
    defaults = {'Sex': '', 'Pclass': 3, 'Age': 30.0, 'SibSp': 0, 'Parch': 0, 'Fare': 15.0}
    d = {**defaults, **passenger}
    sex   = str(d['Sex']).lower()
    pclass= int(d['Pclass'])
    age   = float(d['Age'])
    sibsp = int(d['SibSp'])
    parch = int(d['Parch'])
    fare  = float(d['Fare'])
    fam   = sibsp + parch


    if sex == 'female':
        if pclass == 3:
            if age >= 40:
                return 0
            else:
                if fam >= 3:
                    return 0
                else:
                    return 1
        else:
            return 1
    else:
        if age < 12:
            return 1
        if pclass == 1 and fare >= 50:
            return 1
        if age < 18 and fam >= 2:
            return 1
        return 0

passenger = {
    "PassengerId": 161,
    "Survived": 0,
    "Pclass": 3,
    "Name": "Cribb, Mr. John Hatfield",
    "Sex": "male",
    "Age": 44,
    "SibSp": 0,
    "Parch": 1,
    "Ticket": "371362",
    "Fare": 16.1,
    "Cabin": None,
    "Embarked": "S"
}

pred = predict_survival(passenger)
print(pred, 'Выжил' if pred == 1 else 'Не выжил')
```

```python
import math
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X = new_df.drop(columns=["Survived"]).to_numpy(dtype=float)
y = new_df["Survived"].to_numpy(dtype=float)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

class MyLogisticRegression:
    def __init__(self, lr, iterations):
        self.lr = lr
        self.iterations = iterations
        self.w = []
        self.b = 0

    def sigmoid(self, z):
        return 1 / (1 + math.exp(-z))

    def fit(self, X, y):
        m, n = len(X), len(X[0])
        self.w = [0.0] * n

        for _ in range(self.iterations):
            z = [sum(self.w[j] * X[i][j] for j in range(n)) + self.b for i in range(m)]
            y_pred = [self.sigmoid(z[i]) for i in range(m)]
            errors = [y_pred[i] - y[i] for i in range(m)]

            dw = [sum(errors[i] * X[i][j] for i in range(m)) / m for j in range(n)]
            db = sum(errors) / m

            for j in range(n):
                self.w[j] -= self.lr * dw[j]
            self.b -= self.lr * db

    def predict_proba(self, X):
        z = [sum(self.w[j] * X[i][j] for j in range(len(self.w))) + self.b for i in range(len(X))]
        return [self.sigmoid(val) for val in z]

    def predict(self, X, threshold=0.5):
        return [1 if p >= threshold else 0 for p in self.predict_proba(X)]

my_model = MyLogisticRegression(lr=0.1, iterations=50000)
my_model.fit(X_train.tolist(), y_train.tolist())
pred_my = my_model.predict(X_test.tolist())
prob_my = my_model.predict_proba(X_test.tolist())

sk_model = LogisticRegression()
sk_model.fit(X_train, y_train)
pred_sk = sk_model.predict(X_test)
prob_sk = sk_model.predict_proba(X_test)[:, 1]

def accuracy(y_true, y_pred):
    correct = sum(1 for i in range(len(y_true)) if y_true[i] == y_pred[i])
    return correct / len(y_true) * 100

def mse(y_true, y_pred):
    return sum((y_true[i] - y_pred[i])**2 for i in range(len(y_true))) / len(y_true)

print("Accuracy (test): {:.2f}%".format(accuracy(y_test.tolist(), pred_my)))
print("Accuracy (sklearn, test): {:.2f}%".format(accuracy(y_test, pred_sk)))
print("")
print("MSE: {:.4f}".format(mse(y_test.tolist(), prob_my)))
print("MSE (sklearn, test): {:.4f}".format(mse(y_test, prob_sk)))

print("\nПример предсказаний (тестовые данные):")
for i in range(20):
    print(f"y={int(y_test[i])}, my={pred_my[i]}  (prob={prob_my[i]:.2f}), sk={int(pred_sk[i])} (prob={prob_sk[i]:.2f})")
```
