```python
df['total_bedrooms'] = df['total_bedrooms'].fillna(df['total_bedrooms'].mean())
```

```python
df['ocean_proximity'].value_counts()
```

```python
mapping = {'<1H OCEAN': 1,'INLAND': 2,'NEAR OCEAN': 3,'NEAR BAY':4,'ISLAND':5}
df['ocean_proximity'] = df['ocean_proximity'].map(mapping)
```

```python
df.hist(figsize=(15, 15))
```

```python
import numpy as np

num_df = df.select_dtypes(include=np.number)  # только числовые
corr = num_df.corr()
plt.figure(figsize=(15, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.show()
```

```python
corr_target = df.corr(numeric_only=True)['median_house_value'] \
                .drop('median_house_value') \
                .sort_values(ascending=False)
print("Топ коррелирующих признаков с median_house_value:\n", corr_target.head(5))
```
