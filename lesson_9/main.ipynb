```python
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

dataset = loadtxt('ParisHousing.csv', delimiter=',', skiprows=1)
# split into input (X) and output (y) variables
X = dataset[:, 0:16]
y = dataset[:, 16]

scaler = StandardScaler()
X_norm = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_norm, y, test_size=0.2, random_state=42)

# define the keras model
model = Sequential()
model.add(Dense(32, input_shape=(16,)))
model.add(Dense(16, activation='relu'))
model.add(Dense(1))
# compile the keras model
model.compile(loss='mse', optimizer='adam', metrics=['mae'])

# 
history = model.fit(X_train, y_train, epochs=50, batch_size=10, verbose=1, validation_split=0.2)
mae_per_epoch = history.history['mae']

mae_tuple = tuple(mae_per_epoch)

#   MAE   
min_mae = min(mae_tuple)
min_epoch = mae_tuple.index(min_mae) + 1

print(" MAE:", min_mae)
print("   MAE:", min_epoch)
# evaluate the keras model
_, mae = model.evaluate(X_test, y_test)
print(f"MAE  : {mae:.2f}")
```

```python
import numpy as np

class NeuralNetwork:
    def __init__(self, layers, learning_rate=0.01, epochs=5000):
        self.layers = layers
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.weights = [np.random.randn(layers[i], layers[i+1]) * np.sqrt(2/layers[i])
                        for i in range(len(layers)-1)]
        self.biases = [np.zeros((1, layers[i+1])) for i in range(len(layers)-1)]

        #  Adam,      , - 
        self.m_w = [np.zeros_like(W) for W in self.weights]
        self.v_w = [np.zeros_like(W) for W in self.weights]
        self.m_b = [np.zeros_like(b) for b in self.biases]
        self.v_b = [np.zeros_like(b) for b in self.biases]
        #  Adam
        self.beta1 = 0.9   #     ()
        self.beta2 = 0.999 #     ( )
        self.eps = 1e-8    #     

    @staticmethod
    def relu(x):
        return np.maximum(0, x)

    @staticmethod
    def relu_derivative(x):
        return (x > 0).astype(float)

    def learn(self, X_train, y_train):
        for epoch in range(self.epochs):
            #  
            activations = [X_train] #  
            z_values = []#    (z = XW + b)

            #    
            for i, (W, b) in enumerate(zip(self.weights, self.biases)):
                z = np.dot(activations[-1], W) + b # z = XW + b
                z_values.append(z)
                if i == 0:#
                    activations.append(z)
                elif i < len(self.weights) - 1: #
                    activations.append(self.relu(z))
                else: # 
                    activations.append(z)

            # 
            m = X_train.shape[0]
            error = activations[-1] - y_train
            mse = np.mean(error**2)
            mae = np.mean(np.abs(error))

            if epoch % 100 == 0 or epoch == self.epochs - 1:
                print(f" {epoch}: MSE = {mse:.4f}, MAE = {mae:.4f}")

            #  
            delta = error
            grads_w = []#   
            grads_b = []#   
            for i in reversed(range(len(self.weights))):
                grad_w = np.dot(activations[i].T, delta) / m #   
                grad_b = np.mean(delta, axis=0, keepdims=True) #   
                grads_w.insert(0, grad_w)
                grads_b.insert(0, grad_b)

                #    
                if i > 0:
                    delta = np.dot(delta, self.weights[i].T)
                    if i > 1:
                        delta *= self.relu_derivative(z_values[i-1])

            # Adam  (   )
            for i in range(len(self.weights)):
                # 
                self.m_w[i] = self.beta1 * self.m_w[i] + (1 - self.beta1) * grads_w[i]
                self.v_w[i] = self.beta2 * self.v_w[i] + (1 - self.beta2) * (grads_w[i] ** 2)
                m_hat_w = self.m_w[i] / (1 - self.beta1 ** (epoch + 1))
                v_hat_w = self.v_w[i] / (1 - self.beta2 ** (epoch + 1))
                self.weights[i] -= self.learning_rate * m_hat_w / (np.sqrt(v_hat_w) + self.eps)

                # 
                self.m_b[i] = self.beta1 * self.m_b[i] + (1 - self.beta1) * grads_b[i]
                self.v_b[i] = self.beta2 * self.v_b[i] + (1 - self.beta2) * (grads_b[i] ** 2)
                m_hat_b = self.m_b[i] / (1 - self.beta1 ** (epoch + 1))
                v_hat_b = self.v_b[i] / (1 - self.beta2 ** (epoch + 1))
                self.biases[i] -= self.learning_rate * m_hat_b / (np.sqrt(v_hat_b) + self.eps)

    def predict(self, X):
        activ = X #     
        for i, (W, b) in enumerate(zip(self.weights, self.biases)):
            z = np.dot(activ, W) + b
            if i == 0 or i == len(self.weights) - 1:
                activ = z
            else:
                activ = self.relu(z)
        return activ

    def evaluate(self, X, y):
        y_pred = self.predict(X)
        mae = np.mean(np.abs(y_pred - y))
        return mae

```

```python
X = dataset[:, 0:16]
y = dataset[:, 16].reshape(-1, 1)

scaler_X = StandardScaler()
scaler_y = StandardScaler()
X_norm = scaler_X.fit_transform(X)
y_norm = scaler_y.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X_norm, y_norm, test_size=0.2, random_state=42)

nn = NeuralNetwork(layers=[16, 32, 16, 1], learning_rate=0.01, epochs=1700)
nn.learn(X_train, y_train)

y_pred_norm = nn.predict(X_test)
y_pred = scaler_y.inverse_transform(y_pred_norm)
y_test_orig = scaler_y.inverse_transform(y_test)

mae = np.mean(np.abs(y_pred - y_test_orig))
print(f"MAE  : {mae:.4f}")
```
